```python
import pandas as pd
from sqlalchemy import create_engine
from config import pw
db_path='data/SQL_Challenge.sql'
engine = create_engine(f'postgresql://postgres:{pw}@localhost:5432/SQL_Challenge')
conn = engine.connect()
import matplotlib.pyplot as plt
```

```python
salary_df=pd.read_sql('select salary from salary',conn)
```

```python
plt.hist(salary_df['salary'],bins=10)
plt.ylabel('Index')
plt.xlabel('Salary')
plt.title('Salary Histogram')
```

```python
salary_title=pd.read_sql('select title, round(avg(salary),2) from salary join title using (emp_no) group by title',conn)
salary_title.columns=['Title','Average Salary']
```

```python
salary_title.plot('Title','Average Salary',kind='barh',legend=None)

plt.xlabel('Average Salary')
```
